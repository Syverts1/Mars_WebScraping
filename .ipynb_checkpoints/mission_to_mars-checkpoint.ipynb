# NASA Mars News

```python
url = "https://mars.nasa.gov/news/"
response = requests.get(url)
soup = BeautifulSoup(response.text,'html.parser')

print(soup.prettify())
```

```python
#find news titles within parsed html
title = soup.find_all('div',class_="content_title")

news_title=[]

for item in title:
    contents = item.a.text
    news_title.append(contents.strip('\n'))

print(news_title)
```

```python
#find paragraph text within parsed html
desc = soup.find_all('div', class_="rollover_description_inner")
                      
news_p=[]

for item in desc:
    contents = item.text
    news_p.append(contents.strip('\n'))

print(news_p)
```

# JPL Mars Space Images - Featured Image

```python
#create function that will initialize chrome as the chosen browser to run splinter
def init_browser():
    executable_path = {"executable_path": "chromedriver.exe"}
    return Browser("chrome", **executable_path)
```

```python
#create a scraping function to get JPL's featured space image
def scrape():
    browser = init_browser()

    base_url = "https://www.jpl.nasa.gov"
    search_url = base_url + "/spaceimages/?search=&category=Mars"
    browser.visit(search_url)
    time.sleep(1)

    html = browser.html
    soup = BeautifulSoup(html, "html.parser")

    main_img = soup.find('article', class_="carousel_item").get('style')

    rel_url = ""
    d= False
    base_url = "https://www.jpl.nasa.gov"


    for i in main_img:
        if(i=="'" and d== False):
            d = True
        elif(i=="'" and d==True):
            d = False
        elif(i!="'" and d==True):
            rel_url = rel_url + i

    image_url = base_url + rel_url
    browser.quit()
    return image_url
```

```python
featured_image_url = scrape()

print(featured_image_url)
```

# Mars Weather

```python
#Get Mars Weather
#pull information from mars weather twitter page
url_weather = "https://twitter.com/marswxreport?lang=en"

response_weather = requests.get(url_weather)
```

```python
soup_weather = BeautifulSoup(response_weather.text,'html.parser')

print(soup_weather.prettify())
```

```python
#find paragraph text within parsed html
tweets = soup_weather.find_all('p', class_="TweetTextSize TweetTextSize--normal js-tweet-text tweet-text")
                      
print(tweets[0])
```

```python
weather_p=[]

for item in tweets:
    i = item.text
    if i.startswith("Sol") == True:
        contents = item.text
        weather_p.append(contents.strip('\n'))

mars_weather = weather_p[0]
print(mars_weather)
```

# Mars Facts

```python
#Get Mars Facts
#pull information from space facts page
url_facts = "http://space-facts.com/mars/"

response_facts = requests.get(url_facts)

soup_facts = BeautifulSoup(response_facts.text,'html.parser')

print(soup_facts.prettify())
```

```python
#find paragraph text within parsed html
div_contents = soup_facts.find('div',class_='post-content')
                      
print(div_contents)
```

```python
#pull mars description out of post-content div
para = div_contents.find_all('p')
#print(para)

desc={0:['Description:'],1:[ para[1].text]}

facts_df = pd.DataFrame(data=desc)
print(facts_df)
```

```python
profile = div_contents.table

print(profile)
```

```python
profile_contents = pd.read_html("https://space-facts.com/mars/")[0]

profile_contents
```

```python
#append profile to description to get the full list of facts about mars, in one df
mars_facts = facts_df.append(profile_contents,ignore_index=True)

fixed = mars_facts.set_index(0)

fixed = fixed.rename(index=str,columns={1: 'value'})

fixed.index.name = None

mars_facts_html = fixed.to_html()

print(mars_facts_html)
```

# Mars Hemisperes

```python
#pull html contents of the website containing mars hemisphere information
url_hemi = "https://astrogeology.usgs.gov/search/results?q=hemisphere+enhanced&k1=target&v1=Mars"

response_hemi = requests.get(url_hemi)

soup_hemi = BeautifulSoup(response_hemi.text,'html.parser')
```

```python
#find title of each mars hemisphere

#print(soup_hemi.prettify())
h3 = soup_hemi.find_all('h3')
h3_parsed = []

for item in h3:
    h3_parsed.append(item.text)

print(h3_parsed)
```

```python
#find path to the image page, that will be used by splinter to find the exact path to a sample image
img_tag = soup_hemi.find_all('a', class_="itemLink product-item")

#print(img_tag)

img_path=[]

for item in img_tag:
    i = item.get('href')
    img_path.append('https://astrogeology.usgs.gov' + i)

print(img_path)
```

```python
#create a scraping function to get the url to each image shown on the hemispheres page
def scrape_hemi():
    browser = init_browser()

    soup = []

    for url in img_path:
        browser.visit(url)
        time.sleep(1)

        html = browser.html
        soup.append(BeautifulSoup(html, "html.parser").find('div',class_='downloads'))
    browser.quit()
    
    return soup
```

```python
#run function and put into the downloads variable
downloads = scrape_hemi()
```

```python
#print(downloads)
#parse the downloads list to pull the image url
image_url=[]
    
for item in downloads:
    contents = item.a.get('href')
    image_url.append(contents)

print(image_url)
```

```python
#put the title and image url into a list of dictionaries

hemisphere_image_urls = []

for index,item in enumerate(h3_parsed):

    hemisphere_image_urls.append({"title": item , "img_url": image_url[index]})

print(hemisphere_image_urls)
```
